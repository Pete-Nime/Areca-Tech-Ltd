This code cell is responsible for merging data from various dataframes into a single comprehensive sales_data DataFrame. It combines information about sales transactions with details about the products sold, categories, customers, sales employees, and their locations (cities and countries).

Here's a breakdown of the steps:

product_data = pd.merge(products, categories, on='CategoryID', how='left'): This line merges the products DataFrame with the categories DataFrame based on the CategoryID column. It uses a left merge, keeping all products and adding their category names.
sales_data = pd.merge(sales, product_data, on='ProductID', how='left'): This merges the sales DataFrame with the product_data (which now includes product and category info) based on ProductID. This brings product details into the sales records.
sales_data = pd.merge(sales_data, customers, on='CustomerID', how='left'): This merges the sales_data (with sales and product info) with the customers DataFrame based on CustomerID, adding customer details to the sales records.
sales_data = pd.merge(sales_data, employees, left_on='SalesPersonID', right_on='EmployeeID', how='left'): This merges sales_data with the employees DataFrame. It matches SalesPersonID from sales_data with EmployeeID from employees to add employee details to the sales records.
sales_data = pd.merge(sales_data, cities, left_on='CityID_x', right_on='CityID', how='left', suffixes=('_customer', '_city')): This merges sales_data with the cities DataFrame to get city information. It uses CityID_x (the customer's CityID after the merge with customers) from sales_data and CityID from cities. Suffixes are added to distinguish columns with the same name from the left and right dataframes.
sales_data = pd.merge(sales_data, countries, left_on='CountryID', right_on='CountryID', how='left'): This merges sales_data with the countries DataFrame to add country names. It uses the CountryID column (from the cities data) to join with the CountryID in the countries DataFrame.
The print statements and display(sales_data[['FirstName', 'MiddleInitial', 'LastName']].head()) are debugging steps to inspect the DataFrame after merging with customers.
sales_data.head(): Finally, this line displays the first few rows of the resulting sales_data DataFrame to preview the combined data.
In essence, this cell consolidates data from multiple related tables into a single DataFrame, making it easier to perform analysis that requires information from different sources.



```python
import pandas as pd

# Merge products with categories
product_data = pd.merge(products, categories, on='CategoryID', how='left')

# Merge sales with product_data
sales_data = pd.merge(sales, product_data, on='ProductID', how='left')

# Merge with customers
sales_data = pd.merge(sales_data, customers, on='CustomerID', how='left')

# Debugging: Print columns and head of customer name columns after merging with customers
print("Columns of sales_data after merging with customers:", sales_data.columns.tolist())
if 'FirstName' in sales_data.columns:
    print("\nHead of customer name columns after merging with customers:")
    display(sales_data[['FirstName', 'MiddleInitial', 'LastName']].head())
else:
    print("\nCustomer name columns not found in sales_data after merging with customers.")


# Merge with employees
sales_data = pd.merge(sales_data, employees, left_on='SalesPersonID', right_on='EmployeeID', how='left')

# Merge cities with customers to get customer location
sales_data = pd.merge(sales_data, cities, left_on='CityID_x', right_on='CityID', how='left', suffixes=('_customer', '_city'))

# Merge countries to get country names
sales_data = pd.merge(sales_data, countries, left_on='CountryID', right_on='CountryID', how='left')

# Final clean-up
# sales_data.drop(columns=['city_id_y', 'city_id_customer'], inplace=True) # This line seems to be incorrect based on the corrected merge

# Preview the final dataset
sales_data.head()
```

This code cell performs several data cleaning and feature engineering steps, primarily focusing on the SalesDate and TotalPrice columns.

sales_data.dropna(subset=['SalesDate'], inplace=True): This line removes rows from the sales_data DataFrame where the SalesDate column has missing (null) values. inplace=True means the changes are made directly to the sales_data DataFrame. This is important to avoid errors when converting SalesDate to datetime format later.
sales_data['SalesDate'] = pd.to_datetime(sales_data['SalesDate']): This line converts the SalesDate column to a datetime format. This allows for easier extraction of date components and time-based analysis.
sales_data['year'] = sales_data['SalesDate'].dt.year: This line creates a new column named year and extracts the year from the SalesDate column using the .dt.year accessor.
sales_data['month'] = sales_data['SalesDate'].dt.month: This line creates a new column named month and extracts the month from the SalesDate column using the .dt.month accessor.
sales_data['year_month'] = sales_data['SalesDate'].dt.to_period('M'): This line creates a new column named year_month and extracts the year and month as a Period object (representing a monthly time period) from the SalesDate column using the .dt.to_period('M') accessor. This format is useful for grouping and analyzing data by month.
sales_data['TotalPrice'] = sales_data['Quantity'] * sales_data['Price'] * (1 - sales_data['Discount']): This line calculates the TotalPrice for each sale. It multiplies the Quantity by the Price and then applies the Discount (by multiplying by 1 - Discount). This overwrites the existing TotalPrice column with the calculated values.
In summary, this cell prepares the sales data for time-based analysis and ensures the TotalPrice is correctly calculated based on quantity, price, and discount.



```python
# Drop rows with missing SalesDate
sales_data.dropna(subset=['SalesDate'], inplace=True)

# Convert 'SalesDate' column to datetime
sales_data['SalesDate'] = pd.to_datetime(sales_data['SalesDate'])

# Extract year and month for time-based analysis
sales_data['year'] = sales_data['SalesDate'].dt.year
sales_data['month'] = sales_data['SalesDate'].dt.month
sales_data['year_month'] = sales_data['SalesDate'].dt.to_period('M')

# Calculate TotalPrice using Quantity, Price, and Discount
sales_data['TotalPrice'] = sales_data['Quantity'] * sales_data['Price'] * (1 - sales_data['Discount'])
```

## Regenerate monthly sales plot

```python
import pandas as pd

# Load necessary dataframes and perform essential merges to create sales_data
try:
    sales = pd.read_csv('/content/drive/MyDrive/Grocery_Database_Project_June_10/sales.csv')
    products = pd.read_csv('/content/drive/MyDrive/Grocery_Database_Project_June_10/products.csv')
    categories = pd.read_csv('/content/drive/MyDrive/Grocery_Database_Project_June_10/categories.csv')
except FileNotFoundError as e:
    print(f"Error loading data file: {e}. Please make sure the CSV files are in the correct directory.")
    raise e

# Merge sales with products to get Price and Discount
sales_data = pd.merge(sales, products[['ProductID', 'Price']], on='ProductID', how='left')

# Merge with categories to get CategoryName (if needed for other analysis, not strictly for this cell)
# sales_data = pd.merge(sales_data, categories, on='CategoryID', how='left') # Not strictly needed for this cell

# Calculate TotalPrice (without Discount as it's missing from products)
sales_data['TotalPrice'] = sales_data['Quantity'] * sales_data['Price']

# Drop rows with missing SalesDate and convert to datetime
sales_data.dropna(subset=['SalesDate'], inplace=True)
sales_data['SalesDate'] = pd.to_datetime(sales_data['SalesDate'])

# Extract year and month for time-based analysis and create year_month
sales_data['year'] = sales_data['SalesDate'].dt.year
sales_data['month'] = sales_data['SalesDate'].dt.month
sales_data['year_month'] = sales_data['SalesDate'].dt.to_period('M')


# Group by date (monthly total sales)
monthly_sales = sales_data.groupby('year_month')['TotalPrice'].sum().reset_index()
monthly_sales['year_month'] = monthly_sales['year_month'].astype(str)

# Set date index for forecasting
monthly_sales.index = pd.to_datetime(monthly_sales['year_month'])

# Drop unused column
monthly_sales = monthly_sales.drop(columns='year_month')

# Preview
display(monthly_sales.tail(10))
```

Customer Segmentation (Clustering)


Feature Engineering for Customers


This code cell performs customer segmentation by grouping customers based on their purchasing behavior using KMeans clustering. It calculates key metrics for each customer, prepares the data for clustering, applies the clustering algorithm, and then shows the resulting distribution of customers into different segments.

##### Here is a breakdown of the code:

Aggregate data per customer:
customer_data = sales_data.groupby('CustomerID').agg({...}): This line groups the sales_data DataFrame by CustomerID and then applies aggregation functions to specific columns:
'TotalPrice': 'sum': Calculates the sum of TotalPrice for each customer (total amount spent).
'ProductID': 'count': Counts the number of products purchased by each customer (transaction count).
'SalesDate': ['min', 'max']: Finds the earliest (min) and latest (max) purchase dates for each customer.
The result is stored in the customer_data DataFrame.
Flatten multi-level column names:
customer_data.columns = ['total_spent', 'transaction_count', 'first_purchase', 'last_purchase']: The aggregation step creates multi-level column names (e.g., ('TotalPrice', 'sum')). This line simplifies these column names to single-level, more descriptive names.
Recency calculation:
latest_purchase_date = sales_data['SalesDate'].max(): Finds the latest overall sales date in the entire sales_data DataFrame.
customer_data['recency'] = (latest_purchase_date - customer_data['last_purchase']).dt.days: Calculates the 'recency' for each customer, which is the number of days between the latest overall purchase date and the customer's last purchase date.
Drop date columns:
customer_data = customer_data.drop(columns=['first_purchase', 'last_purchase']): Removes the original first_purchase and last_purchase date columns as they are no longer needed after calculating recency.
Normalize data:
from sklearn.preprocessing import StandardScaler: Imports the StandardScaler class.
scaler = StandardScaler(): Creates an instance of the scaler.
customer_scaled = scaler.fit_transform(customer_data): This scales the features (total_spent, transaction_count, recency) in customer_data. Scaling is important for KMeans as it is sensitive to the magnitude of features. It fits the scaler to the data and then transforms the data.
KMeans clustering:
from sklearn.cluster import KMeans: Imports the KMeans class.
kmeans = KMeans(n_clusters=4, random_state=42): Creates an instance of the KMeans model, specifying 4 clusters and setting a random_state for reproducibility.
customer_data['cluster'] = kmeans.fit_predict(customer_scaled): This line fits the KMeans model to the scaled customer data and predicts the cluster label for each customer. A new column 'cluster' is added to the original customer_data DataFrame containing the assigned cluster ID (0, 1, 2, or 3) for each customer.
View cluster distribution:
display(customer_data['cluster'].value_counts()): This line counts how many customers were assigned to each cluster and displays the counts.
In summary, this cell prepares customer data based on their purchasing activity, scales the data, applies the KMeans algorithm to group customers into 4 segments, and shows how many customers fall into each segment

```python
# Aggregate data per customer
customer_data = sales_data.groupby('CustomerID').agg({
    'TotalPrice': 'sum',
    'ProductID': 'count',
    'SalesDate': ['min', 'max']
})

# Flatten multi-level column names
customer_data.columns = ['total_spent', 'transaction_count', 'first_purchase', 'last_purchase']

# Recency = days since last purchase
# Need to make sure sales_data['SalesDate'].max() is calculated from the same data as customer_data['last_purchase']
latest_purchase_date = sales_data['SalesDate'].max()
customer_data['recency'] = (latest_purchase_date - customer_data['last_purchase']).dt.days

# Drop date columns
customer_data = customer_data.drop(columns=['first_purchase', 'last_purchase'])

# Normalize data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
customer_scaled = scaler.fit_transform(customer_data)

# KMeans clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['cluster'] = kmeans.fit_predict(customer_scaled)

# View cluster distribution
display(customer_data['cluster'].value_counts())
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd # Import pandas here too

# Load necessary dataframes
try:
    sales = pd.read_csv('/content/drive/MyDrive/Grocery_Database_Project_June_10/sales.csv')
    products = pd.read_csv('/content/drive/MyDrive/Grocery_Database_Project_June_10/products.csv')
except FileNotFoundError as e:
    print(f"Error loading data file: {e}. Please make sure the CSV files are in the correct directory.")
    raise e

# Merge sales with products to get ProductName
sales_data = pd.merge(sales, products[['ProductID', 'ProductName']], on='ProductID', how='left')

# Step 1: Select Top 100 Products by frequency
top_products = sales_data['ProductName'].value_counts().head(100).index

# Step 2: Filter sales_data to include only those top products
filtered_data = sales_data[sales_data['ProductName'].isin(top_products)].copy() # Use .copy() to avoid SettingWithCopyWarning

# Step 3: Create basket format: each transaction's product list from filtered data
basket = (
    filtered_data
    .groupby(['TransactionNumber', 'ProductName'])['Quantity']
    .sum()
    .unstack()
    .fillna(0)
)

# Step 4: Convert to binary (1/0)
basket = basket.map(lambda x: 1 if x > 0 else 0)

print(f"Original number of products: {len(sales_data['ProductName'].unique())}")
print(f"Number of products after filtering to top 100: {len(top_products)}")
print(f"Original number of transactions: {sales_data['TransactionNumber'].nunique()}")
print(f"Number of transactions after filtering products: {basket.shape[0]}") # Number of rows in the basket after filtering


# Generate frequent itemsets
# Note: min_support might need adjustment depending on the number of transactions after filtering
frequent_items = apriori(basket, min_support=0.01, use_colnames=True)

# Generate rules
rules = association_rules(frequent_items, metric="lift", min_threshold=1)

# Sort rules by confidence
rules.sort_values(by='confidence', ascending=False).head(10)
```
